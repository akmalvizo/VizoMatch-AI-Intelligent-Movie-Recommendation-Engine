```python
# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# now we convert it into preprocessing format like
# ['Action', 'Adventure', 'Fantasy', 'Science Fiction']
# ast.literal_eval -> It's designed to handle simple Python data structures like lists, dictionaries, and basic types.
```

```python
import ast
def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i['name'])
  return L
```

```python
# Now genres colum store specific values
movies_new['genres'] = movies_new['genres'].apply(convert)
```

```python
movies_new.head(2)
```

```python
# Now keywords colum store specific values
movies_new['keywords'] = movies_new['keywords'].apply(convert)
```

```python
movies_new.head(2)
```

```python
# Now we fetch (Main/First) three movie hero names
def convert3(obj):
  L = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      L.append(i['name'])
      counter +=1
    else:
      break
  return L
```

```python
movies_new['cast'] = movies_new['cast'].apply(convert3)
```

```python
movies_new.head(2)
```

```python
# Now we find director name from crew table
def fetch_director(obj):
  L = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      L.append(i['name'])
      break
  return L
```

```python
movies_new['crew'] = movies_new['crew'].apply(fetch_director) # Pass the function object, not a string
```

```python
movies_new.head(2)
```

```python
# Now we convert string of overview column into list
movies_new['overview'] = movies_new['overview'].apply(lambda x:x.split())
```

```python
movies_new.head(2)
```

```python
# This code remove all blank spaces from these colums
movies_new['genres'] = movies_new['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies_new['keywords'] = movies_new['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies_new['cast'] = movies_new['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies_new['crew'] = movies_new['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies_new.head(2)
```

```python
# Now we concatenate or merge last 4 columns into a new column tag
movies_new['tag'] = movies_new['overview'] + movies_new['genres'] +movies_new['cast'] +movies_new['crew']
```

```python
movies_new.head(2)
```

```python
# Now we make new data frame by using preprocessing
new_df = movies_new[['movie_id', 'title', 'tag']]
```

```python
new_df.head(2)
```

```python
# Now we convert list into string
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df.head(2)
```

```python
# this code convert similer words into one word like( love, loving, loved =love)
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Now we write a streaming function to stem the words mean it convert (loved, loving into love ,,etc,,)
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))  # Append stemmed word to the list
  return " ".join(y)  # Return the joined stemmed string after processing all words
```

```python
new_df['tag'] =new_df['tag'].apply(stem)
```

```python
new_df['tag'][0]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
new_df.head(2)
```

```python
# This code is used for extract (5000) words and remove all stop words of english(is, it, the, to, etc)
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# Now we apply upper code on our new_df datasets tag column
cv.fit_transform(new_df['tag']).toarray().shape
```

```python
# this code convert all string data of tag into array vectors
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]]. title)
  return
```

```python
recommend('Batman Begins')
```
